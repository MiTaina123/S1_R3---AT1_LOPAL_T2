```python
n1 = float(input("Digite o primeiro número: "))
n2 = float(input("Digite o segundo número:  "))
n3 = float(input("Digite o terceiro número: "))

lista = [n1, n2, n3]

maior = max(lista)
menor = min(lista)

print("O maior número é:", maior)
print("O menor número é:", menor)
```

```python
n = int(input("Quantos termos da série de Fibonacci você deseja? "))
a,b = 1, 1
cont = 0

while cont < n:
  print(a, end=" ")
  a, b = b, a + b
  cont += 1
```

```python
while True:
    nome = input("Nome (mín. 3 caracteres): ")
    if len(nome) > 3: break

while True:
    idade = int(input("Idade (0-150): "))
    if 0 <= idade <= 150: break

while True:
    salario = float(input("Salário (maior que 0): "))
    if salario > 0: break

while True:
    sexo = input("Sexo ('f' ou 'm'): ").lower()
    if sexo in ['f', 'm']: break

while True:
    estado_civil = input("Estado Civil ('s', 'c', 'v', 'd'): ").lower()
    if estado_civil in ['s', 'c', 'v', 'd']: break

print("Dados validados com sucesso!")
```

```python
num = int(input("Digite um número inteiro: "))
mult = 0

for i in range(2, num):
  if num % i == 0:
    mult += 1

if mult == 0 and num >1:
  print(f"{num} é primo.")
else:
  print(f"{num} não é primo.")
```

```python
num = int(input("Digite um número para calcular o fatorial: "))
fatorial = 1

for i in range(1, num + 1):
    fatorial *= i

print(f"O fatorial de {num} é {fatorial}.")
```

```python
L = [5, 7, 2, 9, 4, 1, 3]

print(f"a) Tamanho: {len(L)}")
print(f"b) Maior: {max(L)}")
print(f"c) Menor: {min(L)}")
print(f"d) Soma: {sum(L)}")
L.sort()
print(f"e) Crescente: {L}")
L.sort(reverse=True)
print(f"f) Decrescente: {L}")
```
